## Import Library

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from qiskit import *
from qiskit import QuantumCircuit
from qiskit.visualization import plot_state_qsphere, plot_histogram, plot_bloch_multivector, array_to_latex
from qiskit_aer import AerSimulator

from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit_ibm_provider.job import job_monitor
```

## Simple Circuit

```python
circuit = QuantumCircuit(1, 1)

initial_state = [0.6, 0.8] # [probility of 0, probability of 1]
circuit.initialize(initial_state, 0)

circuit.measure(0, 0)

circuit.draw(output='mpl', initial_state=True)
```

## Run on Simulator

```python
circuit = QuantumCircuit(1)

initial_state = [0.6, 0.8] # [probility of 0, probability of 1]
circuit.initialize(initial_state, 0)

circuit.h(0)
circuit.h(0)

circuit.draw(initial_state=True, output='mpl')
```

```python
simulator = AerSimulator()
circuit.save_statevector()
job = simulator.run(circuit)
job_monitor(job)
result = job.result()
final_state = result.get_statevector()
```

## Visualize the final_state/statevector

```python
print(final_state)
array_to_latex(final_state)
```

```python
plot_bloch_multivector(final_state)
```
